```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

# Contoh pemanggilan
bfs(graph, 'A')
```

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

# Contoh pemanggilan
dfs(graph, 'A')
```

```python
from collections import deque


# 1. Membuat Graf Baru dan Menampilkannya
print("--- Tugas 1: Graf Baru ---")
my_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['B', 'C', 'E'],
    'E': ['C', 'D']
}

def print_graph(graf):
    for node in graf:
        print(f"{node} -> {graf[node]}")

print_graph(my_graph)
print("\n")


# 2. Implementasi BFS dan DFS Asli

print("--- Tugas 2: Implementasi BFS dan DFS Asli ---")

def bfs_original(graph, start):
    visited = set()
    queue = deque([start])
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

def dfs_original(graph, start, visited=None):
    if visited is None:
        visited = set()
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs_original(graph, neighbor, visited)

print("Hasil BFS dari 'A': ", end="")
bfs_original(my_graph, 'A')
print("\nHasil DFS dari 'A': ", end="")
dfs_original(my_graph, 'A')
print("\n\n")


# 3. Modifikasi BFS Mengembalikan List

print("--- Tugas 3: Modifikasi BFS (Return List) ---")

def bfs_return_list(graph, start):
    visited = set()
    queue = deque([start])
    urutan_kunjungan = []

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            urutan_kunjungan.append(vertex)
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

    return urutan_kunjungan

hasil_bfs_list = bfs_return_list(my_graph, 'A')
print(f"List urutan kunjungan BFS: {hasil_bfs_list}\n")



# 4. Fungsi find_path Menggunakan DFS
print("--- Tugas 4: Fungsi find_path (DFS) ---")

def find_path(graph, start, end, path=None):
    if path is None:
        path = []

    path = path + [start]

    if start == end:
        return path

    if start not in graph:
        return None

    for node in graph[start]:
        if node not in path:
            newpath = find_path(graph, node, end, path)
            if newpath:
                return newpath

    return None

jalur = find_path(my_graph, 'A', 'E')
print(f"Jalur dari A ke E: {jalur}\n")



# 5. Fungsi is_connected

print("--- Tugas 5: Fungsi is_connected ---")

def get_all_visited_dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    if start not in visited:
        visited.add(start)
        for neighbor in graph[start]:
            get_all_visited_dfs(graph, neighbor, visited)
    return visited

def is_connected(graph):

    if not graph:
        return True
    start_node = list(graph.keys())[0]

    visited_nodes = get_all_visited_dfs(graph, start_node)


    return len(visited_nodes) == len(graph)

status_koneksi = is_connected(my_graph)
print(f"Apakah graf saling terhubung? {status_koneksi}")
```
