```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None # Wajib ada untuk Double Linked List

class LinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # PENYISIPAN TENGAH (Double)
    def addBetween(self, mid_node, newdata):
        if mid_node is None:
            return

        NewNode = Node(newdata)
        # 1. Atur tangan kanan NewNode ke depan
        NewNode.nextval = mid_node.nextval
        # 2. Atur tangan kiri NewNode ke mid_node
        NewNode.prevval = mid_node
        # 3. Hubungkan balik dari node depan ke NewNode
        if mid_node.nextval is not None:
            mid_node.nextval.prevval = NewNode
        # 4. Hubungkan mid_node ke NewNode
        mid_node.nextval = NewNode

    # HAPUS DENGAN KATA KUNCI
    def deleteNode(self, key):
        curr = self.headval

        # Cari node yang datanya cocok dengan key
        while curr is not None:
            if curr.dataval == key:
                break
            curr = curr.nextval

        if curr is None: # Data tidak ditemukan
            return

        # Jika yang dihapus adalah HEAD
        if curr == self.headval:
            self.headval = curr.nextval

        # Atur tali next (node sebelumnya lompat ke node setelahnya)
        if curr.nextval is not None:
            curr.nextval.prevval = curr.prevval

        # Atur tali prev (node setelahnya menunjuk balik ke node sebelum curr)
        if curr.prevval is not None:
            curr.prevval.nextval = curr.nextval

# --- EKSEKUSI ---
n1 = Node("januari")
n2 = Node("Februari")
n3 = Node("maret")

Li = LinkedList()
Li.headval = n1
# Sambungkan Double Linked List awal
n1.nextval = n2
n2.prevval = n1
n2.nextval = n3
n3.prevval = n2

print("--- Sebelum Sisip ---")
Li.listprint()

print("\n--- Setelah Sisip 'Maret_Baru' setelah Februari ---")
Li.addBetween(n2, "Maret_Baru")
Li.listprint()

print("\n--- Setelah Hapus 'januari' ---")
Li.deleteNode("januari")
Li.listprint()
```

```python
class Simpul:
    def __init__(ini, data = None):
        ini.data = data
        ini.berikutnya = None

class LinkedList:
    def __init__(ini):
        ini.kepala = None

    def listprint(ini):
        prints = ini.kepala
        while prints is not None:
            print(prints.data)
            prints = prints.berikutnya

    def TambahAwal(ini , DataBaru):
        DataNew = Simpul(DataBaru)
        DataNew.berikutnya = ini.kepala
        ini.kepala = DataNew

    def TambahTengah(ini , Tengah_data , DataBaru ):
        if Tengah_data is None:
            print("data tengahnya kosong")
            return
        DataNews = Simpul(DataBaru)
        DataNews.berikutnya = Tengah_data.berikutnya
        Tengah_data.berikutnya = DataNews

    def TambahAkhir(ini , DataBaru):
        DataNew = Simpul(DataBaru)
        if ini.kepala is None:
            ini.kepala = DataNew
            return
        last = ini.kepala
        while(last.berikutnya):
            last = last.berikutnya
        last.berikutnya = DataNew


    def HapusNode(ini, KunciHapus):
        sementara = ini.kepala # Kita mulai dari lokomotif (kepala)

        # Skenario 1: Bagaimana kalau node yang mau dihapus ternyata adalah kepala
        if sementara is not None:
            if sementara.data == KunciHapus:
                ini.kepala = sementara.berikutnya
                sementara = None
                return

        # Skenario 2: node yang dicari ada di tengah atau akhir
        sebelumnya = None
        while sementara is not None:
            if sementara.data == KunciHapus:
                break
            sebelumnya = sementara
            sementara = sementara.berikutnya

        # Skenario 3: Sudah jalan sampai ujung, tapi datanya nggak ketemu
        if sementara == None:
            print(f"Data '{KunciHapus}' tidak ditemukan!")
            return
        sebelumnya.berikutnya = sementara.berikutnya
        sementara = None

n1 = Simpul("maret")
n2 = Simpul("april")
n3 = Simpul("may")

li = LinkedList()
li.kepala = n1
li.kepala.berikutnya = n2
n2.berikutnya = n3

print("=== SEBELUM DIHAPUS ===")
li.TambahAwal("awal")
li.TambahTengah(li.kepala, "Tengah")
li.TambahAkhir("last")
li.listprint()

print("\n=== SETELAH MENGHAPUS 'maret' ===")
li.HapusNode("maret")
li.listprint()

print("\n=== SETELAH MENGHAPUS 'awal' (kepala) ===")
li.HapusNode("awal")
li.listprint()
```
