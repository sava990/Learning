```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

# --- Fungsi Traversal dari Gambar Ketiga ---

def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

# --- Membangun Tree (Dari kode sebelumnya) ---

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]

for value in data_list:
    root = bst.insert(root, value)

# --- Menampilkan Output ---

print("Inorder Traversal:")
inorder(root)

print("\n\nPreorder Traversal:")
preorder(root)

print("\n\nPostorder Traversal:")
postorder(root)
print() # Tambahan print kosong agar baris di terminal rapi
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

# --- Fungsi Pencarian dari Gambar Keempat ---
def search(node, key):
    if node is None or node.data == key:
        return node
    if key < node.data:
        return search(node.left, key)
    return search(node.right, key)

# --- Membangun Tree (Dari kode sebelumnya) ---
bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]

for value in data_list:
    root = bst.insert(root, value)

# --- Uji Pencarian (Dari Gambar Keempat) ---
key = 60
result = search(root, key)
if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")

key = 25
result = search(root, key)
if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")
```

```python
class Node:

    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root


def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)


def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')


def search(node, key):
    if node is None or node.data == key:
        return node
    if key < node.data:
        return search(node.left, key)
    return search(node.right, key)


npm_3_digit = 129
npm_2_digit = 29
tanggal_lahir = 10
bulan_lahir = 12


ascii_huruf_1 = ord('M') # Hasilnya 77
ascii_huruf_2 = ord('u') # Hasilnya 117


angka_bebas_1 = 99
angka_bebas_2 = 88


data_list = [
    npm_3_digit,
    tanggal_lahir,
    bulan_lahir,
    ascii_huruf_1,
    ascii_huruf_2,
    angka_bebas_1,
    angka_bebas_2
]


bst = BinarySearchTree()
root = None


for value in data_list:
    root = bst.insert(root, value)

print(f"Data awal yang dimasukkan: {data_list}\n")


print("--- Hasil Traversal ---")
print("Inorder Traversal:")
inorder(root)

print("\n\nPreorder Traversal:")
preorder(root)

print("\n\nPostorder Traversal:")
postorder(root)
print("\n")


print("--- Hasil Pencarian ---")

# Skenario 1: Mencari 2 digit terakhir NPM
hasil_cari_1 = search(root, npm_2_digit)
if hasil_cari_1:
    print(f"Nilai {npm_2_digit} (2 Digit NPM) DITEMUKAN dalam tree.")
else:
    print(f"Nilai {npm_2_digit} (2 Digit NPM) TIDAK DITEMUKAN.")

# Skenario 2: Mencari angka yang pasti tidak ada di list (misalnya angka 404)
angka_hilang = 404
hasil_cari_2 = search(root, angka_hilang)
if hasil_cari_2:
    print(f"Nilai {angka_hilang} DITEMUKAN dalam tree.")
else:
    print(f"Nilai {angka_hilang} TIDAK DITEMUKAN.")
```
